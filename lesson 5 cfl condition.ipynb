```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
def linearconvec(sigma,nx):
    dx = 2./(nx-1)
    dt = sigma*dx
    nt=20
    
    x = np.linspace(0.,2.,nx)
    
    u = np.ones(nx)
    u0 = np.ones(nx)
    u[0.5/dx:1./dx]=2
    u0[0.5/dx:1./dx]=2
    
    for n in range(1,nt):
        un = np.copy(u)
        u[1:] = un[1:]*(1-(dt/dx)*(un[1:]-un[0:-1]))
    plt.title('Sigma = %0.2f'%sigma)
    plt.plot(x,u0,label='Time Zero')
    plt.plot(x,u,label='Time One')
    plt.legend(loc='best')
    plt.ylim(0.0,2.5)
    print ('dt/dx = '),dt/dx
```

```python
linearconvec(0.5,100)
```

```python
linearconvec(0.7,100)
```
